```python
# error 10 pix smaller to 10 pix larger
# size 25 to 400 px
error = np.arange(-10,11,1)
sz = np.arange(25,401,1)
```

```python
# size,error permutations
X,Y = np.meshgrid(sz,error)
```

```python
X.shape,Y.shape
```

```python
X
```

```python
Y
```

```python
pred_area = (X+Y)**2
```

```python
act_area = X**2
```

```python
_ = np.stack((pred_area,act_area),axis=0)
```

```python
inter = _.min(axis=0)
```

```python
dice = 2*(inter/(pred_area+act_area))
```

```python
%matplotlib notebook
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, dice, 300)
ax.set_xlabel('size')
ax.set_ylabel('error')
ax.set_zlabel('dice');
```
